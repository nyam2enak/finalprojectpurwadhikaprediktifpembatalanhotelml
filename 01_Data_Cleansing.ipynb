## Analisis Bisnis Menggunakan Confusion Matrix

|  | Prediksi: Tidak Batal (0) | Prediksi: Batal (1) |
|---|--------------------------|---------------------|
| **Aktual: Tidak Batal (0)** | True Negative (TN) | False Positive (FP) |
| **Aktual: Batal (1)** | False Negative (FN) | True Positive (TP) |

### False Positive (FP)

Model memprediksi akan batal, padahal tidak batal. Dampak bisnisnya :

- Hotel mungkin melakukan overbooking karena mengira kamar akan kosong
- Bisa menyebabkan double booking atau ketidaknyamanan tamu (komplain citra hotel jadi jelek)
- Tetapi ada nilai positifnya, setidaknya pemesanan kamar tidak dibatalkan sehingga hotel masih mendapatkan pemasukan

### False Negative (FN)

Model memprediksi tidak akan batal, padahal dibatalkan. Dampak bisnisnya :

- Hotel kehilangan potensi pendapatan karena kamar kosong tidak sempat dijual kembali
- Kerugian langsung karena tidak ada tamu yang datang
- Kondisi ini membuat kerugian bagi hotel semakin besar dimana pihak operasional sudah menambah karyawan dan logistik tetapi ada pembatalan, terutama bila pembatalan bersifat mendadak

### True Positive (TP)

- Dengan mengetahui probabilitas pembatalan, pihak hotel bisa mencari cara untuk membuat calon pelanggan tetap memesan kamar
- Diharapkan dengan cara persuasif dapat memberikan pemasukan bagi hotel, walaupun ada biaya yang harus dikeluarkan seperti promo atau peningkatan pelayanan
- Memberikan gambaran tentang karakter dari pesanan pelanggan, sehingga bisa digunakan untuk memprediksi kebutuhan pelanggan di masa yang akan datang

## Strategi Memprediksi Menggunakan Machine Learning

Membangun model prediksi pembatalan reservasi hotel yang mampu mengidentifikasi kemungkinan pembatalan dengan akurasi tinggi untuk mendukung strategi overbooking dan pengelolaan pendapatan. 

### Tujuan Spesifik

- Memprediksi apakah reservasi akan dibatalkan (is_canceled) berdasarkan data historis pelanggan dan booking
- Mencapai nilai **F2 Score** yang lebih tinggi. Metrik ini memang lebih menekankan pada recall, sehingga cocok untuk kasus di mana False Negative (FN) perlu diminimalkan. Recall mengukur seberapa banyak kasus pembatalan yang berhasil diprediksi sebagai pembatalan, dan dianggap False Negative (FN) paling merugikan.
- Menggunakan data historis yang sudah tersedia dan model seperti XGBoost, LightGBM, Linear Regression, Decision Tree, atau Random Forest

### F2 Score

Rumus umum dari F-beta score adalah:

$$
F_\beta = (1 + \beta^2) \cdot \frac{\text{Precision} \cdot \text{Recall}}{(\beta^2 \cdot \text{Precision}) + \text{Recall}}
$$

Untuk **F2 Score**, nilai $\beta = 2$, sehingga rumusnya menjadi:

$$
F_2 = \frac{5 \cdot \text{Precision} \cdot \text{Recall}}{4 \cdot \text{Precision} + \text{Recall}}
$$

---

## Ekspektasi Penggunaan Metrik F2 Score Secara Bisnis

### 1. Mengurangi Kerugian Akibat Pembatalan Mendadak (False Negative)

- Karena F2 lebih menekankan recall, model akan lebih sensitif terhadap kasus pembatalan
- Membantu mengidentifikasi lebih banyak potensi pembatalan, sehingga hotel bisa mengambil tindakan preventif seperti:
  - Menawarkan diskon agar pelanggan tidak membatalkan
  - Menyiapkan strategi overbooking yang lebih aman
  - Menyesuaikan alokasi staf dan logistik

### 2. Meningkatkan Efisiensi Operasional

Dengan prediksi yang lebih akurat terhadap pembatalan, hotel bisa:

- Menghindari pemborosan sumber daya (tenaga kerja, makanan, kebersihan)
- Mengoptimalkan penjadwalan staf dan persiapan kamar
- Mengurangi biaya tetap yang tidak menghasilkan pendapatan

### 3. Memaksimalkan ROI dari Model Prediktif

- Dengan mengurangi FN, kerugian $500 per kasus bisa ditekan, sehingga net benefit dan ROI meningkat

| **is_peak_season**     | Indikator apakah booking pada musim ramai (contoh: Jun–Aug, Dec)        | Mengukur dampak musim terhadap pola reservasi                                              | Peak season → cancel rate lebih rendah, relevan untuk revenue management |
| **season**             | Kategori musim dari `arrival_date_month` (Winter, Spring, Summer, Fall) | Mengelompokkan pola musiman booking                                                        | Berguna untuk analisis tren, strategi pricing, dan segmentasi pasar     |


```python
#CELL 18 Menambah Fitur Baru
df_prep = df.copy()

# 1. Total nights stay
if {'stays_in_weekend_nights','stays_in_week_nights'}.issubset(df_prep.columns):
    df_prep['total_nights'] = df_prep['stays_in_weekend_nights'] + df_prep['stays_in_week_nights']
    print("   ✅ Created 'total_nights'")

# 2. Total people
people_cols = ['adults','children','babies']
existing_people_cols = [c for c in people_cols if c in df_prep.columns]
if existing_people_cols:
    df_prep['total_people'] = df_prep[existing_people_cols].sum(axis=1)
    print("   ✅ Created 'total_people'")

# 3. Weekend ratio
if {'stays_in_weekend_nights','total_nights'}.issubset(df_prep.columns):
    df_prep['weekend_ratio'] = df_prep['stays_in_weekend_nights'] / (df_prep['total_nights'] + 1e-8)
    print("   ✅ Created 'weekend_ratio'")

# # 4. Room type match
# if {'reserved_room_type','assigned_room_type'}.issubset(df_prep.columns):
#     df_prep['room_type_match'] = (df_prep['reserved_room_type'] == df_prep['assigned_room_type']).astype(int)
#     print("   ✅ Created 'room_type_match'")

# 5. Booking complexity
complexity_features = ['booking_changes','total_of_special_requests']
existing_complexity = [c for c in complexity_features if c in df_prep.columns]
if existing_complexity:
    df_prep['booking_complexity'] = df_prep[existing_complexity].sum(axis=1)
    print("   ✅ Created 'booking_complexity'")

# 6. Customer reliability
if {'previous_cancellations','previous_bookings_not_canceled'}.issubset(df_prep.columns):
    total_prev = df_prep['previous_cancellations'] + df_prep['previous_bookings_not_canceled']
    df_prep['customer_reliability'] = df_prep['previous_bookings_not_canceled'] / (total_prev + 1e-8)
    print("   ✅ Created 'customer_reliability'")

# 7. Lead time categories
if 'lead_time' in df_prep.columns:
    df_prep['lead_time_category'] = pd.cut(
        df_prep['lead_time'],
        bins=[-1, 7, 30, 90, 365, df_prep['lead_time'].max()],
        labels=['Very_Short','Short','Medium','Long','Very_Long']
    )
    print("   ✅ Created 'lead_time_category'")

# 8. Seasonal indicators
if 'arrival_date_month' in df_prep.columns:
    season_map = {
        'December':'Winter','January':'Winter','February':'Winter',
        'March':'Spring','April':'Spring','May':'Spring',
        'June':'Summer','July':'Summer','August':'Summer',
        'September':'Fall','October':'Fall','November':'Fall'
    }
    df_prep['season'] = df_prep['arrival_date_month'].map(season_map)
    peak_months = ['June','July','August','December']
    df_prep['is_peak_season'] = df_prep['arrival_date_month'].isin(peak_months).astype(int)
    print("   ✅ Created 'season' and 'is_peak_season'")

# 10. Is Loss Cancellation
# Jika deposit_type adalah Non Refund dan is_canceled 1 adalah tidak termasuk loss cancellation karena sudah bayar, lainnnya 0
# if {'deposit_type','is_canceled'}.issubset(df_prep.columns):
#     df_prep['is_loss_cancellation'] = np.where(
#         (df_prep['deposit_type'] == 'Non Refund') & (df_prep['is_canceled'] == 1),
#         0,
#         np.where(df_prep['is_canceled'] == 1, 1, 0)
#     )
#     print("   ✅ Created 'is_loss_cancellation'")
```
